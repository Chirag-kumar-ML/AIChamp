### Task 2:Extracting text from the above PDFs Files  and store them in a CSV.

```python
lin = []
total_check = 0
for i in file:
    lines=""
    with pdfplumber.open(i) as pdf:
        pages = pdf.pages
        for page in pdf.pages:
            text = page.extract_text()
            for line in text.split('\n'):
                lines+=line
        lin.append(lines)        
        
            
            
            
```

```python
from pandas import DataFrame
df = DataFrame (lin,columns=['Profile_Information'])
df.to_csv(r'persons.csv')
```

### Task 3: Code that extracts most frequent words and essential words

```python
import collections
import matplotlib.pyplot as plt
import re
import nltk
from nltk.stem import PorterStemmer
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
STOP_WORDS = stopwords.words("english")
print("111")
for i in range(50):
    a=df.iloc[i, 1]
# Instantiate a dictionary, and for every word in the file, 
# Add to the dictionary if it doesn't exist. If it does, increase the count.
    wordcount = {}# To eliminate duplicates, remember to split by punctuation, and use case demiliters.
    for word in a.split():
        word= str(word).lower()
        word = word.replace(".","")
        word = word.replace(",","")
        word = word.replace(":","")
        word = word.replace("\"","")
        word = word.replace("!","")
        word = word.replace("Â","")
        word = word.replace("â˜","")
        word = word.replace("-â","")
        word = word.replace("2020â","")
        word = word.replace("2018â","")
        word = word.replace(" ","")
        word = word.replace("-","")
        
        word = word.replace(",000,000", "m").replace(",000", "k").replace("′", "'").replace("’", "'")\
                           .replace("won't", "will not").replace("cannot", "can not").replace("can't", "can not")\
                           .replace("n't", " not").replace("what's", "what is").replace("it's", "it is")\
                           .replace("'ve", " have").replace("i'm", "i am").replace("'re", " are")\
                           .replace("he's", "he is").replace("she's", "she is").replace("'s", " own")\
                           .replace("%", " percent ").replace("₹", " rupee ").replace("$", " dollar ")\
                           .replace("€", " euro ").replace("'ll", " will")
        word = re.sub(r"([0-9]+)000000", r"\1m", word)
        word = re.sub(r"([0-9]+)000", r"\1k", word)
            
        if word not in STOP_WORDS:
            if word not in wordcount:
                wordcount[word] = 1
            else:
                wordcount[word] += 1# Print most common word
    n_print = 5#int(input("How many most common words to print: "))
    print("\nOK. The {} most common words are as follows\n".format(n_print))
    word_counter = collections.Counter(wordcount)
    for word, count in word_counter.most_common(n_print):
        print(word, ": ", count)# Close the file
    
     #Draw a bar chart
#     lst = word_counter.most_common(n_print)
#     df = pd.DataFrame(lst, columns = ['Word', 'Count'])
#     df.plot.bar(x='Word',y='Count')
    
    
```
